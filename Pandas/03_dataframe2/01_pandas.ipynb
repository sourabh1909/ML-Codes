#### Long vs wide data

1) wide (tall) format is where we have a [single row] for every data point with multiple columns to hold te values of various attributes.


2) long format is where for each data point we have as many rows as the number of attribute and each row contains the value of a perticulae attributes for a given data point.

- melt (wide -- long)

- pivot  (long -- wide)

```python
# both data format will give same info
# but in analyzing may differ 

# for that we see next two functions

# melt  ->  wide format ko long format  mhanje row extend karayche
# pivot -> long ko wide mai convert kar deta hai


# this is example for wide to long
pd.DataFrame({'cse':[120],
              'ese':[150],
              'mech':[50]}).melt(var_name='branch',value_name='students')
```

```python
# this example for long to wide

pd.DataFrame(
    {
        'branch':['cse','ese','mech'],
        '2020':[100,150,80],
        '2021':[120,160,90],
        '2022':[190,80,50]
    }
).melt(id_vars=['branch'],var_name='year',value_name='students')

# id_vars madhe je taknar tyache column madhe conversion honar nahi.
```
