```python
marks = {
    'maths' : 67,
    'english' : 85,
    'science' : 89,
    'hindi':78
}
# key value pairs
# name is provided in marks
marks_series = pd.Series(marks,name='Sourabh ke marks')
marks_series
```

```python
# using indexing
marks_series[3] = 78
marks_series
```

```python
# what if index is not given
marks_series['sst'] = 150
marks_series
# this will add new value in your data series
```

```python
# slicing
marks_series[1:3] = [100,100]
marks_series
```

```python
marks_series[[0,1]] = [0,10]
marks_series
```

Boolean Indexing on series

find no of 50's and 100's scored by kohli   
find number of ducks   
count number of day when i had more than 200 subd a day  
find actors who have dine more than 20 movies   

```python
# find number of 50's and 100's scored by kohli

# here boolean array ko hum index ke form mai lakar vahase data retrive karenge
vk = pd.read_csv('kohli_ipl.csv',usecols=['runs']).squeeze(True)
vk[((vk>= 50) & (vk<=100))|(vk >= 100)].count()
```
