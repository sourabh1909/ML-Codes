```python
# To deal with missing values
temp = pd.Series([1,2,3,np.nan,5,6,np.nan,8,np.nan,10])
print(temp)

temp[(temp.isnull() == 0)]
```

```python
temp = pd.Series([1,2,3,np.nan,5,6,np.nan,8,np.nan,10])
x = int(input("Enter the number for missing values."))
# once we edited value we can not changes further
temp[temp.isnull()] = x
temp
```
