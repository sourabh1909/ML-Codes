```python
import numpy as np
import pandas as pd
```

Series from lists

```python
country = ['india','pakistan','USA','Nepal']

pd.Series(country)  # series is class inside pandas
                    # ek object tayar kela 
```

```python
#  index     values
# series object mai do chije hoti hai
# data type in object
# if in pandas the dtype is object then 99% have probability that, that object is string

runs = [13,24,56,8,4,100]
pd.Series(runs,dtype='int32')
```

```python
# custom index name
marks = [67,57,89,100]
sub = ['maths','physics','english','hindi']

pd.Series(marks,index= sub) #by default the data is int64 we can change its by dtype
```

```python
import matplotlib.pyplot as plt

marks = [35, 40, 42, 55, 60, 61, 65, 70, 75, 80]
bins = [30, 40, 50, 60, 70, 80, 90]
plt.hist(marks, bins=bins)
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.title("Marks Distribution")
plt.show()
```

```python
# setting a name
marks = [67,57,89,100]
sub = ['maths','physics','english','hindi']

pd.Series(marks,index= sub,name='Sourabh') #by default the data is int64 we can change its by dtype
```

```python
marks = {
    'maths' : 67,
    'english' : 85,
    'science' : 89,
    'hindi':78
}
# key value pairs
# name is provided in marks
marks_series = pd.Series(marks,name='Sourabh ke marks')
marks_series
```
