```python
# using lists
student_data = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]

# series ssathi jase series attribute use karto tase dataframe sathi datframe attribute use karayche
# tyanantar comma deun columns mhunun to column kasachi info deto te specify karayche
pd.DataFrame(student_data,columns=['iq','marks','package'])
```

```python
# using dicts

student_dict = {
    'iq' :[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,3,0,0]
}

# here column will atomatically fetch by keys of dictonary

students = pd.DataFrame(student_dict)
print(students)
```

6) isnull -- returns boolean value for null is present or not

7) duplicated -- shows how many duplicated data
8) rename -- changes the name of column in table
