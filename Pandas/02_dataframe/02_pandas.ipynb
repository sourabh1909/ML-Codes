Selection Rows from a dataFrame

- iloc -- searches using index positions (last wala not included)
- loc --  searches using index labels  (last wala included)

```python
# multiple rows
movie.iloc[0:5:2]
movie.iloc[0:]
movie.iloc[:5]
movie.iloc[[0,4,5]]  #fancy indexing
# now this is dataFrame
```
