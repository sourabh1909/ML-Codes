```python
point_df = pd.DataFrame({
    '1st point' : [(3,4),(-6,5),(0,0),(-10,1),(4,5)],
    '2nd point': [(-3,4),(0,0),(2,2),(10,10),(1,1)]
})
point_df
```

```python
# important formula 
# euclidean distance

def euclidean(row):
    pt_A = row['1st point']
    pt_B = row['2nd point']
    
    return ((pt_A[0] - pt_B[1])**2 + (pt_A[1] - pt_B[1])**2)**0.5

point_df['distance'] = point_df.apply(euclidean,axis=1)
# purn dataframe var apply karaycha ahe mhunun axis =1 pass kele

point_df

# jevha aapan apply he series barobar kam kato tevha column var kam karto 
# jevha aapan apply he dataframe barobar kam karto tevha row var kam karto

```
