@ step to solve these type of question 
  
    1. groupby on perticular column which you want   
    2. desired column to fetch answer   
    3. use aggregate function  
    4. uses sort_value(ascending=False)  

@ Groupby Attribute and methods

  1) find total number of groups  -> len (no of groups make) 
  2) find item in each group -> size  
  3) first()/ last() -> nth item   
  4) get_group -> vs filtering   
  5) groups  
  6) describe
  7) sample
  8) nunique()  -> will give count of total unique column

```python
# all imp methods and functions

len(movie.groupby('Genre'))  # (14)
movie['Genre'].nunique()   # (14)
movie.groupby('Genre').size()
movie['Genre'].value_counts()

genres = movie.groupby('Genre')

genres.first()
genres.last()
genres.nth(7) # index is 8 

genres.get_group('Horror')  # means it will return all movies present in that genres
movie[movie['Genre'] == 'Fantasy']  # same result 
# but get_group is do it in better way

genres.groups  # give a dict of index with perticular groupby

genres.describe()  # same as series and dataframe

genres.sample()
# genres.sample(2,replace=True)
```

```python
# aggregate function
# flexibility
#  by passing dict

genres.agg(
    {
        'Runtime' : 'mean',
        'IMDB_Rating': 'mean',
        'No_of_Votes': 'sum',
        'Metascore' : 'min'
    }
)
```

```python
# by passing list
genres.agg(['min','max','sum'])
# it consist more information
```

```python
# adding both list + dict
genres.agg(
    {
        'Runtime' : ['mean','max'],
        'IMDB_Rating': ['mean','min'],
        'No_of_Votes': ['sum','min'],
        'Metascore' : ['min','min']
    }
)
```

```python
# looping in groups
for group,data in genres:
    print(group)
    
# find the highest rated movie of each genre
# df = pd.DataFrame(columns=movie.columns)
# for group,data in genres:
#     df = df.append(data[data['IMDB_Rating']==data['IMDB_Rating'].max()])
    
# df

# find the highest rated movie of each genre
df_best = movie.loc[movie.groupby('Genre')['IMDB_Rating'].idxmax()]
df_best
```

```python
# split (apply) combine -> strategy
# apply -> builtin function

genres.apply(min).head(5)
```

```python
# find number of movies starting with 'A' for each group 
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
    
genres.apply(foo).head(5)

```

```python
# find ranking od each movie in the group according to IMDB_Score
def rank_movies(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group
genres.apply(rank_movies).head(5)
```

```python
# find normalized IMDB rating group 
#  (x- xmin) / (x - xmax) 
def normal(group):
    group['norm_rating'] =  (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group

genres.apply(normal).head(4)

# it will return old dataframe with this new column
```

```python
ipl = pd.read_csv('deliveries.csv') 
ipl.head(5)
```

```python
ipl.shape
```

```python
# find the 10 batsman with most runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
size = ipl[ipl['batsman_runs'] == 6]

size.groupby('batsman')['batsman'].count().sort_values(ascending=False)
size.groupby('batsman')['batsman'].value_counts().sort_values(ascending=False)
```

```python
# find the batman with most number of 4s and 6s in last 5 overs

temp_df = ipl[ipl['over'] > 15]
overs=temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]
overs.groupby('batsman')['batsman'].value_counts().sort_values(ascending=False).head(1).index[0]

# pahila fitering karaychi nantar 
```

```python
# # find V kolhi's record againset all teams
batter = ipl[ipl['batsman'] == 'V Kohli']
batter.groupby('bowling_team')['batsman_runs'].sum().reset_index().head(5)
```

```python
# create a function that can return the highest score of any batsman

def high(batsman):
    temp_df = ipl[ipl['batsman'] == batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False).head(1).values[0]
     
high('DA Warner')
```
