```python
# used for deep learning
# softmax function
# softmax chya array chi sum 1 aste

def softmax(arr):
    if type(arr) != np.ndarray:
        raise TypeError("Required Numpy Array")
    s= np.sum( np.exp(arr))
    return np.exp(arr) / s
softmax(np.arange(5))
```

```python
def vertical(*args):
    for i in args:
        if type(i) is not np.ndarray:
            raise TypeError("Required numpy array")
    return np.vstack(args)

a = np.arange(10).reshape(2,-1)
print("a:",a)
b = np.arange(10).reshape(2,-1)
print("b:",b)

print(vertical(a,b))
c = np.random.random((2,5))
print("c:",c)
vertical(a,b,c)
```
