Indexing and Slicing

```python
import numpy as np

a1 = np.arange(10)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)
```

```python
# indexing
a1[len(a1)-1]
```

```python
a2[2,2]
a2[2][2]
a2
# both are valid cases
```

```python
a1
```

```python
a1[2:5:2] #slicing
```

Iterating


```python
a2
```

```python
a2[0,:]
a2[1,:2]
a2[:,2]
# a2[row,:col]
# a2[row,:]
# a2[:,col]
```

```python
a2[1:,1:3]
a2
```

```python
a2[0::2,1::2]
```

```python
a2[1,::3]
```

```python
a2[:2:,1::]
```

```python
a3 = np.arange(27).reshape(3,3,3)
a3
```

```python
a3[1,:,2]
```

```python
a3[2,1:,1:]
```

```python
a3[::2,:1:,::2]
```

Iterating

```python
a1
for i in a1:
    print(i)
```

```python
a2
for i in a2:
    for j in i:
        print(j)
```

```python
a3
for i in a3:
    for j in i:
        for k in j:
            print(k)
```

```python
for i in np.nditer(a2):
    print(i)
```

Stacking

```python
# horizaontaly
a4 = np.arange(4).reshape(2,2)
a5 = np.arange(4,8).reshape(2,2)

np.hstack((a4,a5))
np.vstack((a4,a5))
```

Spliting

```python
a6 = np.arange(8).reshape(4,2)
np.hsplit(a6,2)
np.vsplit(a6,2)
```
