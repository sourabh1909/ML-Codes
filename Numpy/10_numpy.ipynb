```python
# DAVT LAB No. 4

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def simulate_poisson_dist(lambda_,size):
    data = np.random.poisson(lam=lambda_,size=size)
    return data

result = simulate_poisson_dist(8,1000)
print(result[:20])
```

```python
def calculate_statistics(data):
    mean = np.mean(data)
    std = np.std(data)
    median = np.median(data)
    min = np.min(data)
    max = np.max(data)
    return mean,std,median,min,max

mean,std,median,min,max = calculate_statistics(result)

print("mean :",mean)
print("standard deviation :",std)
print("median : ",median)
print("min : ",min)
print("max : ",max)
```

```python
# histogram is for data visulization 
def plot_hist(data):
    plt.hist(data)
    plt.xlabel("number of events : ")
    plt.ylabel("freq : ")
    plt.title('Poisson Distrubution')
    plt.show()

plot_hist(result)
```
