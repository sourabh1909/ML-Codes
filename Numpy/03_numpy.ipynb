#Broadcasting

The term broacasting describes how numpy treats array with diff shapes during arithmatic operations. The smaller array is "Broadcasting " across the larger array so that they have  compatible shapes.


```python
# same shape 
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)

print(a)
print(b)
print(a+b)
```

```python
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(12).reshape(4,3)
b = np.arange(3)

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(3)

print(a)
print(b)

# print(a+b)
```

```python
a = np.arange(6).reshape(2,3)
b = np.arange(1,2)

print(a)
print(b)

print(a+b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(4).reshape(4,1)

print(a)
print(b)

print(a+b)
```

```python
a = np.array([1])
b = np.arange(4)

print(a)
print(b)
print(a+b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)

print(a)
print(b)
print(a+b)
```
