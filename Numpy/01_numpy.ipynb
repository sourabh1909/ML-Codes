Array Operations


```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
```

```python
# scalar operations(+,-,*,/)
a1**2
```

```python
#relational
a2>=10
a2[a2 >=10]
```

```python
# vector operations
# a1*a2
np.dot(a1,a2)
```

Array Functions

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
# min/max/sum/prod
# 0 -> col and 1 -> row
np.min(a1,axis=0)
np.min(a1,axis=1)
```

```python
# mean/median/std/var
np.mean(a1,axis=1)
np.median(a1,axis=0)
```

```python
#trigo function
np.sin(a1[2])
```
