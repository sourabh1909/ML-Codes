##### Subplots 
when we do operations on multiple columns
then we won't that graph will plot on single graph
for that we have to use subplot that plots each graph on new graph plot

```python
# there are diff way for plotting graph in matplotlib
# the function name is plt.subplots() 

batter.head(5)
plt.scatter(batter['avg'],batter['strike_rate'])
```

```python
fig,ax = plt.subplots(figsize=(15,6))

ax.scatter(batter['avg'],batter['strike_rate'],color='red')
ax.set_title('Graph')
ax.set_xlabel('Avg')
ax.set_ylabel('strike_rate')

fig.show()
```

```python
# o create multiple plots using subplots we have to nrows and ncols parameter

fig , ax = plt.subplots(nrows=2,ncols=1,sharex=True)
# sharex is used for to show both graphs have common x axis name
ax[0].scatter(batter['avg'],batter['strike_rate'],label='batter_name',color='red')
ax[1].scatter(batter['avg'],batter['runs'],label='name')
ax[0].legend()
ax[1].legend()

ax[0].set_title('Info Of Batters')
ax[1].set_title('Info Of Batters')
ax[1].set_xlabel('Avg')
```

```python
# o create multiple plots using subplots we have to nrows and ncols parameter

fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
# sharex is used for to show both graphs have common x axis name
ax[0,0].scatter(batter['avg'],batter['strike_rate'],label='batter_name',color='red')
ax[0,1].scatter(batter['avg'],batter['runs'],label='name')
ax[1,0].scatter(batter['avg'],batter['runs'],label='name')
ax[1,1].scatter(batter['avg'],batter['runs'],label='name')
ax[0,0].legend()
ax[1,0].legend()

ax[0,0].set_title('Info Of Batters')
ax[1,0].set_title('Info Of Batters')
ax[1,0].set_xlabel('Avg')
```

```python
# o create multiple plots using subplots we have to nrows and ncols parameter
# this is very preferd method 

fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))
# sharex is used for to show both graphs have common x axis name
ax[0,0].scatter(batter['avg'],batter['strike_rate'],label='batter_name',color='red')
ax[0,1].scatter(batter['avg'],batter['runs'],label='name')
ax[1,0].hist(batter['avg'],label='avg of batsman')
ax[1,1].hist(batter['runs'],label='runs of batsman')
ax[0,0].legend()
ax[0,1].legend()
ax[1,0].legend()
ax[1,1].legend()

ax[0,0].set_title('Info Of Batters')
ax[1,0].set_title('avg Of Batters')
ax[1,0].set_xlabel('Avg')
```

```python
# this is littel bit diff method

fig = plt.figure()

# 2 rows 2 column tyachya madhe ha pahila graph
ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'],batter['strike_rate'],label='batter_name',color='red')

# 2 rows 2 column tyachya madhe ha dusara graph
ax2 = fig.add_subplot(2,2,2)
ax2.hist(batter['runs'])

# 2 rows 2 column tyachya madhe ha tisara graph
ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['avg'])
```

- 3D Surface Plots  
-- this very useful in machine learning

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
z = xx**2 + yy**2 
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.subplot(projection='3d')

ax.plot_surface(xx,yy,z,cmap='jet')

# to show colorbar we pass that plot in a varible then fig.colorbar(variable)
p = ax.plot_surface(xx,yy,z,cmap='jet')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(10,10))
ax = plt.subplot(projection='3d')

ax.plot_surface(xx,yy,z,cmap='jet')
```
