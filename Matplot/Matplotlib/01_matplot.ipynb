```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# 2d line graph
# when to use

# Bivariate Analysis  (agar do column ke uper sath mai analysis karne ke liye)
# categorical -> numerical and numerical -> numerical
# time series data ke uper hum 2d line graph use karte hai
```

```python
# plotting 2d data
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
plt.grid()
# plt.plot(x_axis,y_axis)
```

```python
# from pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman.head(2)

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

# categorical data -> numrical data
# x_axis -> y_axis
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')

plt.title('Rohit sharma verses Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

# color is hidden parameter in plt.plot()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dotted')
```

```python
plt.scatter(batsman['index'],batsman['V Kohli'],marker='*',linewidths=5)
```

### plt.Scatter 
##### used for Bivariate simple function
##### numerical vs numrical 
##### use case - finding correlation

```python
x = np.linspace(-10,10,100)
y = 10*x + 3 + np.random.randint(0,110,100)

plt.scatter(x,y)
```

```python
# scatter on proper dataset
batsman = pd.read_csv('batter.csv')
batsman.head(5)

df = batsman.head(50)

plt.scatter(df['strike_rate'],df['avg'],marker='+',linewidths=10,color = 'red')
plt.title('Avg and strike rate analysis')
plt.xlabel('Strike rate')
plt.ylabel('avg')
plt.legend()

# plt.plot(df['strike_rate'],df['avg'])

# diff between scatter and plot is that plot connect all points
# where as scatter is used to denotes point 

# but in real life scatter is much slower than plot function
```

## Bar Chat

```python
# uses
# bivariable analysis
# numerical vs categorical
# use case - Aggregate analysis of groups


children = [10,20,40,10,30]
color = ['red red red red red red red red','blue blue v blue blue','pink pink ink [pink]','yellow','black']

# plt.barh(color,children,color='red')
plt.bar(color,children,color='red',label='liked color')
plt.legend()
 
plt.xticks(rotation=90)
plt.show()
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
# to show multiple columns side by side
# we have to use width and custom sizing

plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,label='2015 runs')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,label='2016 runs')
plt.bar(np.arange(df.shape[0])+0.2,df['2016'],width=0.2,label='2017 runs')

plt.legend()
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
# stacked bar chart

plt.bar(df['batsman'],df['2015'],label='2015 runs')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016 runs')
plt.bar(df['batsman'],df['2017'],bottom=df['2015'] + df['2016'],label='2017 runs')

plt.legend()
```
